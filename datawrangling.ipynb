```python
import pandas as pd
import numpy as np
import re
import pickle
```

```python
hdi = pd.read_csv("Human Development Index - Full.csv")
```

```python
colnames = list(hdi.columns)
```

```python
removecols = []
for item in colnames:
        try:
                if int(item.split(" ")[-1][1:-1]) < 2001:
                        removecols.append(item)
        except:
                continue
```

```python
hdi = hdi.drop(removecols, axis = 1)
```

```python
groupcols = list(hdi.columns)
```

```python
varcol = []
for item in groupcols:
        if re.fullmatch("\(.*\)", item.split(" ")[-1]):
                varcol.append(" ".join(item.split(" ")[:-1]))
```

```python
varcol = np.array(varcol)
```

```python
truevarcol = []
for item in varcol:
        if sum(varcol == item) == 21:
                truevarcol.append(item) 
```

```python
lowerboundbins = np.arange(2001, 2022, 3)
upperboundbins = np.arange(2003, 2022, 3)
```

```python
for item in truevarcol:
        for i in range(len(lowerboundbins)):
                hdi[f'Percent Change, {item} ({lowerboundbins[i]}-{upperboundbins[i]})'] = ((hdi[f'{item} ({upperboundbins[i]})'] - hdi[f'{item} ({lowerboundbins[i]})']) / hdi[f'{item} ({lowerboundbins[i]})']) * 100
```

```python
hdicols = list(hdi.columns)
```

```python
removecols = []
for item in hdicols:
        if item[:15] != "Percent Change," and item != "Country":
                removecols.append(item)
```

```python
hdi = hdi.drop(removecols, axis = 1)
```

```python
finalcols = np.array([])
```

```python
for item in list(hdi.columns):
        finalcols = np.append(finalcols, " ".join(item.split(" ")[:-1]))
```

```python
finalcols = np.unique(finalcols)[1:]
```

```python
hdi = hdi.dropna(axis = 0, subset = ["Percent Change, Human Development Index (2001-2003)", "Percent Change, Human Development Index (2004-2006)", "Percent Change, Human Development Index (2007-2009)", "Percent Change, Human Development Index (2010-2012)", "Percent Change, Human Development Index (2013-2015)", "Percent Change, Human Development Index (2016-2018)", "Percent Change, Human Development Index (2019-2021)"])
```

```python
tabledict = {}
for i in range(1, hdi.shape[0] + 1):
        countrytable = hdi.head(i).tail(1)
        countryname = countrytable["Country"].iloc[0]
        countrytable = countrytable.drop("Country", axis = 1)
        newtable = pd.DataFrame(columns = ["2001-2003", "2004-2006", "2007-2009", "2010-2012", "2013-2015", "2016-2018", "2019-2021"], index = finalcols)
        for column in countrytable:
                newtable[column.split(" ")[-1][1:-1]][" ".join(column.split(" ")[:-1])] = countrytable[column].iloc[0]
        tabledict[countryname] = newtable.transpose()
```

```python
with open("countryframes.pickle", 'wb') as handle:
        pickle.dump(tabledict, handle, protocol = pickle.HIGHEST_PROTOCOL)
```

```python
with open("hdi.pickle", 'wb') as handle:
        pickle.dump(hdi, handle, protocol = pickle.HIGHEST_PROTOCOL)
```
